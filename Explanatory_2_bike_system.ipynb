## Investigation Overview
> The overall goal of this presentation is to show the distribution of the user types of the bike share system, how they are distributed across genders and what benefit certain user type enjoy.

### Summary
- The percentage of users that are subscribers is more than that of customers. 89% of the users are subscribers while 11% are customers.
- The two user types (subscribers and customers) use the bike share for all trips differently. The difference is so significant. Only subscribers use the bkes share for all trip.
- The subscribers and customers are made up of various gender types. There are more male users than female and other.
- The gender with the lowest number of users has the highest average duration in seconds.

## Dataset Overview

> The Fordgo bike trip dataset includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.
> The dataset was downloaded from the Project page of Udacity Data analyst nanodegree program.
> The dataset will be explored programmatically to discover insights needed for the bike-sharing system to make data-driven decisions.


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
#load the dataset into dataframe
df = pd.read_csv('201902-fordgobike-tripdata.csv')

#cinvert to object datatype
df['start_station_id'] = df['start_station_id'].astype(object)
df['start_station_latitude'] = df['start_station_latitude'].astype(object)
df['start_station_longitude'] = df['start_station_longitude'].astype(object)
df['end_station_id'] = df['end_station_id'].astype(object)
df['end_station_latitude'] = df['end_station_latitude'].astype(object)
df['end_station_longitude'] = df['end_station_longitude'].astype(object)
df['bike_id'] = df['bike_id'].astype(object)
df['member_birth_year'] = df['member_birth_year'].astype(object)
```

## Distribution of User Type

> A crucial insight from this dataset is that subscribers frequently use the bike-share system than customers. 89% of the bike use is from the subscribers and 11% from the customers. 

> From the chart below, out of 183412 users, 163544 users are subscribers while the remaining are customers.

```python
#create a bar chart with the user_type column
df['user_type'].value_counts().plot(kind='bar', title='Distribution of User Type')
plt.xlabel('User type')
plt.ylabel('Frequency of Users')
plt.xticks(rotation=0);
```

# How the user types use bikes differently.

> The two user types use the bikes differently. Averagely, customers spend more time using the bikes. However, there seems to be a benefit to being a subscriber. Only subscribers use the bike for all trip.

```python
sns.countplot(data=df, x='bike_share_for_all_trip', hue='user_type')
plt.xlabel('Bike share for all or not')
plt.ylabel('Frequency')
plt.title('User type and the use of bike share for all trip');
```

# The distribution of gender across the user types

> The gender that is most popular across the user types is the male gender. This infers that male gender tends to use a bike than the other genders. We also notice that a higher percentage of the male are subscribers.

```python
sns.countplot(data=df, x='user_type', hue='member_gender')
plt.xlabel('User type')
plt.ylabel('Frequency')
plt.title('User type and Gender Type');
```

# How does gender affect the duration of use of bike?

> From the chart below, we see that other gender type has the highest average duration in seconds even with the smallest number of users while Male gender has the lowest average duration in seconds even with the highest number of users.

```python
df.groupby('member_gender')['duration_sec'].mean().plot(kind='bar')
plt.xlabel('Gender Type')
plt.ylabel('Average Duration in seconds')
plt.title('Relationship between gender and duration')
plt.xticks(rotation=0);
```
